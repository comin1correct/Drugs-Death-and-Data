```python
AHBT = 'Affordable_Housing_by_Town_2011-Present.csv'
data = pd.read_csv(AHBT)
data['Town'] = [row.upper() for row in data['Town']]
data.set_index('Town', inplace=True)
del data['Code']

data.head()
```
